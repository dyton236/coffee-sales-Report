```python
import pandas as pd
```

```python
df=pd.read_csv(r"c:\Users\Bhagwati Prasad\Downloads\coffee_sales.csv")
```

```python
df
```

```python
# find the average sale amount per coffee type
df.groupby('coffee_name')['money'].mean()
```

```python
# get all sales that used cash instead of card
df[df['cash_type'] =='cash']
```

```python
# Add a new column "Day_Revenue" showing total sales per date 
df['Day_Revenue'] = df.groupby('Date')['money'].transform('sum')
df.head()
```

```python
# Find the number of sales per month (using Month_name)
df.groupby('Month_name')['money'].count()
```

```python
# Check if there are any duplicate transcation
df.duplicated().sum()
```

```python
# Find the percentage of each coffee type sold
(df['coffee_name'].value_counts(normalize=True)*100).round(2)
```

```python
# Find on which day and time the highest single sale occured
df.loc[df['money'].idxmax()]
```

```python
# Find the top 3 most sold coffee types
df['coffee_name'].value_counts().head(3)
```

```python
# Display correlation between hour_of_day and money
df[["hour_of_day","money"]].corr()
```

```python
# What is the most popular coffee (by count) during Morning vs Afternoon vs Evening?
df.groupby(['Time_of_Day','coffee_name']).size().unstack().fillna(0)
```
